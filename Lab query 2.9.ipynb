```python
import pymysql
from sqlalchemy import create_engine
import pandas as pd
import getpass  
password = getpass.getpass()
```

```python
connection_string = 'mysql+pymysql://root:' + password + '@localhost/sakila'
engine = create_engine(connection_string)

query = '''select COUNT(DISTINCT last_name) from sakila.actor;'''

data = pd.read_sql_query(query, engine)
data
```

```python
query = """SELECT *, 
CASE 
WHEN DATE_FORMAT(rental_date,'%%a') = 'Sat' then 'weekend'
WHEN DATE_FORMAT(rental_date,'%%a') = 'Sun' then 'weekend'
ELSE 'weekday'
END AS day_type
FROM sakila.rental;"""
 
data = pd.read_sql_query(query, engine)
data
```

```python
query = '''SELECT * FROM sakila.film WHERE title like "%%ARMAGEDDON%%";'''

data = pd.read_sql_query(query, engine)
data
```

```python
query = """SELECT title, length FROM sakila.film
GROUP BY LENGTH
ORDER BY LENGTH DESC
LIMIT 10;"""

data = pd.read_sql_query(query, engine)
data
```

```python
query = """SELECT COUNT(film_id), special_features FROM sakila.film
WHERE special_features like '%%Behind the Scenes%%';"""

data = pd.read_sql_query(query, engine)
data
```

```python
query = """SELECT rating, AVG(length) AS film_duration FROM sakila.film
GROUP BY rating 
HAVING film_duration > 120;"""

data = pd.read_sql_query(query, engine)
data
```

```python
query = """SELECT title,length, RANK() OVER (
ORDER BY length ASC
)length_rank FROM sakila.film
WHERE (
length is not null 
and 
length != 0
)
Group by length;"""

data = pd.read_sql_query(query, engine)
data
```
